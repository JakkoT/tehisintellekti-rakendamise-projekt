# Andmetega tutvumine

Selles notebook'is analüüsime toorandmestikku.

## 1. Andmete laadimine ja põhiinfo

```python
import pandas as pd

# Laadime andmed
df = pd.read_csv('andmed/toorandmed.csv')

print(f"Andmepunktide arv (ridade arv): {df.shape[0]}")
print(f"Tunnuste arv (veergude arv): {df.shape[1]}")
print(f"\nAndmestiku dimensioonid: {df.shape}")
```

## 2. Ühe andmerea põhjalik vaatlus

Vaatame 256. ainekursust detailselt läbi.

```python
# Valime 456. rea (456. aine)
valitud_aine = df.iloc[256]

print("=" * 80)
print("256. AINE DETAILNE INFO")
print("=" * 80)
print()

# Kuvame kõik andmed selle aine kohta
for tunnus, vaartus in valitud_aine.items():
    print(f"{tunnus}:")
    print(f"  {vaartus}")
    print()
```

```python
# Kokkuvõte 256. ainest põhitunnuste kohta
print("=" * 80)
print("256. AINE PÕHIINFO KOKKUVÕTE")
print("=" * 80)
print()
print(f"Ainekood: {valitud_aine['code']}")
print(f"Aine nimetus (EN): {valitud_aine['title__en']}")
print(f"Aine nimetus (ET): {valitud_aine['title__et']}")
print(f"Ainepunktid (EAP): {valitud_aine['credits']}")
print(f"Olek: {valitud_aine['state__et']} ({valitud_aine['state__en']})")
print(f"Viimane uuendus: {valitud_aine['last_update']}")
```
